```python
import numpy as np
from PIL import Image, ImageShow
from matplotlib import pyplot as plt
from IPython import display
```

```python
x = np.load('../data/demos_reach_object_no_distractors_48_16x16.npz', allow_pickle=True)
for k in x.keys():
    print(k)
print(len(x['demo_action_sequences']))
```

```python
x_data = x['demo_image_sequences']
print(x_data.shape, x_data.min(), x_data.max())
marker = [0] + list(x['step_marker'])
print([marker[i+1] - marker[i] for i in range(len(marker) -1)])
```

```python
for i in range(3):
    print(x_data[:,:,:,i].min(), x_data[:,:,:,i].max())
```

```python
xint = np.array(x_data * 255, dtype=np.uint8)
imgs = [Image.fromarray(i) for i in xint]
print(len(imgs))
```

```python
for i in range(len(imgs)):
    plt.figure(i)
    plt.imshow(xint[3])
```

```python
imgs[0].save('../data/16x16_48_trajectories.gif', save_all=True, append_images=imgs[1:])
```
